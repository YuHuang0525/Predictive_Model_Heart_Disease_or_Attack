# Latent Factor Model using fastFM

```python
import csv
import sklearn
import scipy
import numpy as np
from sklearn.model_selection import train_test_split
from fastFM import als
```

```python
dataset = []
```

```python
with open('heart_disease_health_indicators_BRFSS2015.csv', newline='') as csvfile:
    read = csv.DictReader(csvfile)
    for row in read:
        dataset.append(row)
```

```python
len(dataset)
```

```python
y = []
for x in dataset:
    y.append(int(float(x['HeartDiseaseorAttack'])))
```

```python
y = [1 if x else -1 for x in y]
```

Split the data into training data and temp_data by the ratio of 70:30, where temp_data will later be splitted into validation set and test set by the ratio of 50:50. Overall the training, validation, and test data are randomly splitted by 70:15:15.

```python
x_train, temp_x, y_train, temp_y = sklearn.model_selection.train_test_split(dataset, y, test_size=0.3, train_size=0.7, random_state=None, shuffle=True)
```

```python
x_valid = temp_x[:len(temp_x)//2]
y_valid = temp_y[:len(temp_y)//2]
x_test = temp_x[len(temp_x)//2:]
y_test = temp_y[len(temp_y)//2:]
```

Initially we only include features that have high correlation with 'HeartDiseaseorAttack', namely, GenHealth, Age, DiffWalk, HighBP, Stroke, PhysHlth, HighChol, Diabetes, Smoker. 

```python
GenHlth = dict()
Age = dict()
DiffWalk = dict()
HighBP = dict()
Stroke = dict()
PhysHlth = dict()
HighChol = dict()
Diabetes = dict()
Smoker = dict()
for x in dataset:
    if x['GenHlth'] not in GenHlth: GenHlth[x['GenHlth']] = len(GenHlth)
    if x['Age'] not in Age: Age[x['Age']] = len(Age)
    if x['DiffWalk'] not in DiffWalk: DiffWalk[x['DiffWalk']] = len(DiffWalk)
    if x['HighBP'] not in HighBP: HighBP[x['HighBP']] = len(HighBP)
    if x['Stroke'] not in Stroke: Stroke[x['Stroke']] = len(Stroke)
    if x['PhysHlth'] not in PhysHlth: PhysHlth[x['PhysHlth']] = len(PhysHlth)
    if x['HighChol'] not in HighChol: HighChol[x['HighChol']] = len(HighChol)
    if x['Diabetes'] not in Diabetes: Diabetes[x['Diabetes']] = len(Diabetes)
    if x['Smoker'] not in Smoker: Smoker[x['Smoker']] = len(Smoker)
```

```python
n_GenHlth = len(GenHlth)
n_Age = len(Age)
n_DiffWalk = len(DiffWalk)
n_HighBP = len(HighBP)
n_Stroke = len(Stroke)
n_PhysHlth = len(PhysHlth)
n_HighChol = len(HighChol)
n_Diabetes = len(Diabetes)
n_Smoker = len(Smoker)
```

```python
X_train = scipy.sparse.lil_matrix((len(x_train), n_GenHlth + n_Age + n_DiffWalk 
                                   + n_HighBP + n_Stroke + n_PhysHlth + n_HighChol + n_Diabetes + n_Smoker))
X_valid = scipy.sparse.lil_matrix((len(x_valid), n_GenHlth + n_Age + n_DiffWalk 
                                   + n_HighBP + n_Stroke + n_PhysHlth + n_HighChol + n_Diabetes + n_Smoker))
X_test = scipy.sparse.lil_matrix((len(x_test), n_GenHlth + n_Age + n_DiffWalk 
                                   + n_HighBP + n_Stroke + n_PhysHlth + n_HighChol + n_Diabetes + n_Smoker))
```

```python
# define a function that helps transform the data into sparse matrix m.
def transform(data, m):
    for i in range(len(data)):
        ghlth = GenHlth[data[i]['GenHlth']]
        age = Age[data[i]['Age']]
        dwalk = DiffWalk[data[i]['DiffWalk']]
        hBp = HighBP[data[i]['HighBP']]
        stroke = Stroke[data[i]['Stroke']]
        phlth = PhysHlth[data[i]['PhysHlth']]
        hchol = HighChol[data[i]['HighChol']]
        dia = Diabetes[data[i]['Diabetes']]
        smo = Smoker[data[i]['Smoker']]
        m[i,ghlth] = 1 # one-hot encoding GenHlth
        m[i,n_GenHlth + age] = 1 # one-hot encoding Age
        m[i,n_GenHlth + n_Age + dwalk] = 1 # one-hot encoding DiffWalk
        m[i,n_GenHlth + n_Age + n_DiffWalk + hBp] = 1 # one-hot encoding HighBP
        m[i,n_GenHlth + n_Age + n_DiffWalk + n_HighBP + stroke] = 1 # one-hot encoding Stroke
        m[i,n_GenHlth + n_Age + n_DiffWalk + n_HighBP + n_Stroke + phlth] = 1 # one-hot encoding PhysHlth
        m[i,n_GenHlth + n_Age + n_DiffWalk + n_HighBP + n_Stroke + n_PhysHlth + hchol] = 1 # one-hot encoding HighChol
        m[i,n_GenHlth + n_Age + n_DiffWalk + n_HighBP + n_Stroke + n_PhysHlth + n_HighChol + dia] = 1 # one-hot encoding diabetes
        m[i,n_GenHlth + n_Age + n_DiffWalk + n_HighBP + n_Stroke + n_PhysHlth + n_HighChol + n_Diabetes + smo] = 1 # one-hot encoding smoker
        
    return m
```

```python
X_train_filled = transform(x_train, X_train)
X_valid_filled = transform(x_valid, X_valid)
X_test_filled = transform(x_test, X_test)
```

```python
fm = als.FMClassification(n_iter=1000, init_stdev=0.1, rank=5, l2_reg_w=0.1, l2_reg_V=0.1, l2_reg=None)
```

```python
fm.fit(X_train_filled, np.array(y_train))
```
